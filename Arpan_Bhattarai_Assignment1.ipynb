## Question 6: List Sanitization and Metrics
Given:
```python
marks = [45, -3, 88, "NA", 102, None, 67]
```

Write a function that:
1. Removes invalid entries  
2. Returns:
   - Cleaned list  
   - Number of invalid entries  
   - Percentage of valid marks  

Return values using a tuple.


````python
def sanitize_marks(marks):
    #empty list to store valid marks
    cleaned_list = []
    
    # counter for invalid entries
    invalid_count = 0

    # loop through each item in the list
    for mark in marks:
        #check if mark is an integer and non-negative
        if isinstance(mark, int) and mark >= 0:
            cleaned_list.append(mark)
        else:
            invalid_count += 1
    
    total_items = len(marks)
    #percentage
    valid_percentage = (len(cleaned_list) / total_items) * 100
    return (cleaned_list, invalid_count, valid_percentage)

marks = [45, -3, 88, "NA", 102, None, 67]
result = sanitize_marks(marks)
cleaned, invalid, percent = result

print("valid marks:", cleaned)
print("invalid count:", invalid)
print("percentage:", percent)
print(result)

````

## Question 8: Prime Number Function
Write `is_prime(n)` and display prime numbers between 1 and 50.


````python
def is_prime(n):
    #0 and 1 are not prime numbers
    if n <= 1:
        return False

    #check divisibility from 2 to n-1
    for i in range(2, n):
        if n % i == 0:
            return False

    #if no divisor is found, number is prime
    return True


#print prime numbers from 1 to 50
for i in range(1, 51):
    #if true then print
    if is_prime(i):
        print(i)
````

## Question 9: Function Composition
Write functions to:
1. Calculate total marks  
2. Calculate average  
3. Classify result (Distinction / Pass / Fail)

Then write a function that combines them and returns:
```
(name, total, average, result)
```

Explain why modular design is beneficial.


````python
#calculate total marks
def calculate_total(marks):
    return sum(marks)

#calculate average marks
def calculate_average(total, count):
    return total / count

#classify result
def classify_result(average):
    if average >= 75:
        return "Distinction"
    elif average >= 40:
        return "Pass"
    else:
        return "Fail"

#combined function using function composition
def student_result(name, marks):
    total = calculate_total(marks)
    average = calculate_average(total, len(marks))
    result = classify_result(average)
    return (name, total, average, result)

marks = [80, 70, 85]
output = student_result("Ram", marks)
print(output)

#modular design is used by dividing the problem into small functions
#this makes the program easy to read, understand, reuse, and maintain
````

## Question 10: Defensive Input Parsing
Write a function that accepts a raw input string in the format:

```
"Ram|101|22|78, 88, 91"
```
Hint: Parse student data from string format: "Name|Roll|Age|Mark1, Mark2, Mark3"

Tasks:
1. Extract name, roll number, age, and marks  
2. Convert them into appropriate data types  
3. Validate age (≥16) and marks (0–100)  
4. Return the result as a dictionary using roll number as the key  

The function must not crash for malformed input.


````python
def parse_student(data):
    try:
        #split the input string using |
        parts = data.split("|")

        #first part is the name
        name = parts[0]

        #second part contains roll number and age
        roll_age = parts[1].split()
        roll = int(roll_age[0])   #convert roll to integer
        age = int(roll_age[1])    #convert age to integer

        #third part contains marks separated by commas
        marks = parts[2].split(",")
        marks = [int(m) for m in marks]   # convert marks to integers

        #check if age is valid
        if age < 16:
            return "age must be 16 or above"

        #check if marks are between 0 and 100
        for m in marks:
            if m < 0 or m > 100:
                return "marks must be between 0 and 100"

        #return result as dictionary with roll number as key
        return {roll: (name, age, marks)}

    except:
        #runs if input format is wrong
        return "invalid input"


#example input
data = "Ram|101 22|78,88,91"

#function call
print(parse_student(data))
````

## Question 11: Type Stability and Conversion
Given the list:
```python
data = [10, "20", 30.5, "abc", None, True]
```

Write a function that:
1. Separates numeric and non-numeric values  
2. Converts numeric values to float  
3. Returns both lists  

Explain in comments why `True` behaves as a numeric value.


````python
def separate_and_convert(data):
    numeric = []       #list to store numeric values
    non_numeric = []   #list to store non-numeric values

    for item in data:
        #check if item is int or float
        #true is also considered numeric in Python
        if isinstance(item, (int, float)):
            numeric.append(float(item))   #convert numeric value to float
        else:
            non_numeric.append(item)      #store non-numeric value

    return numeric, non_numeric

data = [10, "20", 30.5, "abc", None, True]

num_list, non_num_list = separate_and_convert(data)

print("numeric values:", num_list)
print("non-numeric values:", non_num_list)

#True behaves like a number because in Python:
    #True = 1 and False = 0
#Boolean is a subclass of int
````

## Question 13: Set-Based Enrollment Analysis
Given:
```python
python = {"Amit", "Sita", "Hari", "Nita"}
sql = {"Hari", "Ram", "Nita"}
statistics = {"Amit", "Hari", "Gita"}
```

Find:
1. Students enrolled in exactly two subjects  
2. Students enrolled in only one subject  
3. Students enrolled in all subjects


````python
python = {"Amit", "Sita", "Hari", "Nita"}
sql = {"Hari", "Ram", "Nita"}
statistics = {"Amit", "Hari", "Gita"}

#students in all subjects
all_subjects = python & sql & statistics

#students in exactly two subjects
two_subjects = (python & sql) | (python & statistics) | (sql & statistics)
two_subjects = two_subjects - all_subjects

#students in only one subject
one_subject = (python | sql | statistics) - (python & sql) - (python & statistics) - (sql & statistics)

print(all_subjects)
print(two_subjects)
print(one_subject)
````

## Question 14: Dictionary Aggregation Logic
Given:
```python
records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}
```

Write a function that:
1. Computes average per student  
2. Identifies top and bottom performers  
3. Returns results using a single dictionary


````python
records = {
    "Amit": [78, 82, 91],
    "Sita": [88, 79, 85],
    "Hari": [45, 52, 60]
}

def analyze_records(records):
    averages = {}

    #calculate average for each student
    for name in records:
        marks = records[name]
        avg = sum(marks) / len(marks)
        averages[name] = avg

    #find top and bottom performer
    top = max(averages, key=averages.get)
    bottom = min(averages, key=averages.get)

    return {
        "averages": averages,
        "top_performer": top,
        "bottom_performer": bottom
    }


result = analyze_records(records)
print(result)
````

## Question 15: Nested Data Structure Analysis
Given a nested dictionary representing a school's grade book:
```python
school_data = {
    "Class A": {
        "Amit": [78, 82, 91],
        "Sita": [88, 79, 85]
    },
    "Class B": {
        "Hari": [45, 52, 60],
        "Ram": [92, 88, 95]
    }
}
```

Write a function that:
1. Calculates the average marks for each student
2. Finds the top performer in each class
3. Calculates the overall school average
4. Returns a dictionary with:
   - `class_toppers`: {class_name: (student_name, average)}
   - `school_average`: float
   - `student_averages`: {student_name: average}

Explain in comments why nested dictionaries are useful for hierarchical data.

````python
school_data = {
    "Class A": {
        "Amit": [78, 82, 91],
        "Sita": [88, 79, 85]
    },
    "Class B": {
        "Hari": [45, 52, 60],
        "Ram": [92, 88, 95]
    }
}

def analyze_school(data):
    student_avg = {}
    class_toppers = {}
    total_marks = 0
    total_count = 0

    for class_name in data:
        highest_avg = 0
        topper_name = ""

        for student in data[class_name]:
            marks = data[class_name][student]
            avg = sum(marks) / len(marks)

            student_avg[student] = avg
            total_marks += sum(marks)
            total_count += len(marks)

            if avg > highest_avg:
                highest_avg = avg
                topper_name = student

        class_toppers[class_name] = (topper_name, highest_avg)

    school_average = total_marks / total_count

    return class_toppers, school_average, student_avg


# function call
class_toppers, school_avg, student_avg = analyze_school(school_data)

print("Class Toppers:")
for cls in class_toppers:
    name, avg = class_toppers[cls]
    print(cls, "->", name, "(", round(avg, 2), ")")

print("\nStudent Averages:")
for student in student_avg:
    print(student, "->", round(student_avg[student], 2))

print("\nSchool Average:")
print(round(school_avg, 2))

#nested dictionaries are useful because they store hierarchical data
#like school → class → students → marks in an organized way
````
